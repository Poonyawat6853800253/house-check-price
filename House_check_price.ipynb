```python
# ------------------------------
# 1️⃣ โหลดไฟล์ train/test cleaned
# ------------------------------
import pandas as pd

train_cleaned = pd.read_csv('train_cleaned.csv')
test_cleaned = pd.read_csv('test_cleaned.csv')

print("Train shape:", train_cleaned.shape)
print("Test shape:", test_cleaned.shape)

# ------------------------------
# 2️⃣ แยก Features / Target
# ------------------------------
y_train = train_cleaned['SalePrice']
X_train = train_cleaned.drop(columns=['Id', 'SalePrice'])
X_test = test_cleaned.drop(columns=['Id'])

# ------------------------------
# 3️⃣ One-hot encoding (ถ้ามี categorical)
# ------------------------------
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# ให้ columns ของ test match train
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# ------------------------------
# 4️⃣ Train Linear Regression
# ------------------------------
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

# ------------------------------
# 5️⃣ ทำนาย Test Data และสร้าง submission
# ------------------------------
y_test_pred = model.predict(X_test)

submission = pd.DataFrame({
    'Id': test_cleaned['Id'],
    'SalePrice': y_test_pred
})
submission.to_csv('submission.csv', index=False)
print("Submission file created: submission.csv")
submission.head()

# ------------------------------
# 6️⃣ Validation RMSE & MAE (optional)
# ------------------------------
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
model.fit(X_tr, y_tr)
y_val_pred = model.predict(X_val)

rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
mae = mean_absolute_error(y_val, y_val_pred)
print("Validation RMSE:", rmse)
print("Validation MAE:", mae)

# ------------------------------
# 7️⃣ Interactive: ทำนายราคาบ้านจาก 3 feature หลัก
# ------------------------------
# ดูช่วงค่าของ feature หลัก ๆ
print("\nช่วงค่าของ feature หลัก ๆ:")
print(train_cleaned[['LotArea','OverallQual','YearBuilt']].describe())

# ให้ผู้ใช้กรอกค่า
lot_area_input = float(input("กรุณากรอกขนาดพื้นที่บ้าน (LotArea, ft²) [ตัวอย่าง 8000]: "))
overall_qual_input = int(input("กรุณากรอกคุณภาพบ้าน (OverallQual, 1-10) [ตัวอย่าง 7]: "))
year_built_input = int(input("กรุณากรอกปีสร้างบ้าน (YearBuilt) [ตัวอย่าง 2005]: "))

# สร้าง DataFrame สำหรับทำนาย
X_new = pd.DataFrame(0, index=[0], columns=X_train.columns)
X_new['LotArea'] = lot_area_input
X_new['OverallQual'] = overall_qual_input
X_new['YearBuilt'] = year_built_input

# ทำนายราคาบ้าน
pred_price = model.predict(X_new)
print(f"\nราคาบ้านที่ทำนาย: {pred_price[0]:,.2f} บาท")
```

```python
# ============================================================
# House Prices - Ridge + Pipeline (sklearn 1.6.x friendly)
# - Pipeline + ColumnTransformer ป้องกัน leakage
# - OneHotEncoder(handle_unknown='ignore', sparse_output=False)
# - Target log1p ด้วย TransformedTargetRegressor
# - Metrics: RMSE/MAE/RMSLE (คำนวณเอง เลี่ยงปัญหา shadow ของ metrics)
# - 5-fold CV (neg_root_mean_squared_error)
# - สร้าง submission.csv และเซฟโมเดล
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import joblib
import sklearn
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import RidgeCV
from sklearn.compose import TransformedTargetRegressor
from sklearn.model_selection import train_test_split, KFold, cross_val_score

print("scikit-learn version:", sklearn.__version__)
RANDOM_STATE = 42

# ========== Helper metrics (คำนวณเอง เพื่อตัดปัญหา shadow) ==========
def rmse_score(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def mae_score(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    return np.mean(np.abs(y_true - y_pred))

def rmsle_score(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    y_pred = np.clip(y_pred, a_min=0, a_max=None)  # ต้องไม่ติดลบ
    return np.sqrt(np.mean((np.log1p(y_true) - np.log1p(y_pred)) ** 2))

# ========== 1) Load data ==========
train_cleaned = pd.read_csv('train_cleaned.csv')
test_cleaned = pd.read_csv('test_cleaned.csv')
print("Train shape:", train_cleaned.shape)
print("Test shape :", test_cleaned.shape)

# ========== 2) Split X / y ==========
y = train_cleaned['SalePrice']
X = train_cleaned.drop(columns=['Id', 'SalePrice'])
X_test_raw = test_cleaned.drop(columns=['Id'])

# ========== 3) Identify column types ==========
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()
print(f"Numeric cols: {len(num_cols)}, Categorical cols: {len(cat_cols)}")

# ========== 4) Preprocess + Model ==========
numeric_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# sklearn 1.6.x: ใช้ sparse_output แทน sparse
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

categorical_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', ohe)
])

preprocess = ColumnTransformer(
    transformers=[
        ('num', numeric_pipe, num_cols),
        ('cat', categorical_pipe, cat_cols)
    ],
    remainder='drop'
)

ridge = RidgeCV(alphas=np.logspace(-3, 3, 25), cv=5)
reg = TransformedTargetRegressor(
    regressor=ridge,
    func=np.log1p,
    inverse_func=np.expm1
)

pipe = Pipeline(steps=[
    ('preprocess', preprocess),
    ('reg', reg)
])

# ========== 5) Hold-out validation (20%) ==========
X_tr, X_val, y_tr, y_val = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)
pipe.fit(X_tr, y_tr)
y_val_pred = pipe.predict(X_val)

rmse = rmse_score(y_val, y_val_pred)
mae = mae_score(y_val, y_val_pred)
rmsle = rmsle_score(y_val, y_val_pred)

print("\nHold-out validation (20%):")
print(f"- RMSE : {rmse:,.4f}")
print(f"- MAE  : {mae:,.4f}")
print(f"- RMSLE: {rmsle:,.4f}")

# ========== 6) 5-fold Cross-Validation RMSE ==========
cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_scores = cross_val_score(pipe, X, y, cv=cv, scoring='neg_root_mean_squared_error')
print("\n5-fold CV RMSE:")
print(f"- Mean: {-cv_scores.mean():,.4f}  |  Std: {cv_scores.std():,.4f}")

# ========== 7) Fit on full train + submission ==========
pipe.fit(X, y)
y_test_pred = pipe.predict(X_test_raw)
y_test_pred = np.clip(y_test_pred, a_min=0, a_max=None)  # กันค่าติดลบในไฟล์ส่ง

submission = pd.DataFrame({'Id': test_cleaned['Id'], 'SalePrice': y_test_pred})
submission.to_csv('submission.csv', index=False)
print("\nSubmission file created: submission.csv")
print(submission.head())

# ========== 8) Save model ==========
joblib.dump(pipe, 'house_price_pipeline.joblib')
print("Saved model: house_price_pipeline.joblib")

# ========== 9) Top coefficients (optional) ==========
try:
    feature_names = pipe.named_steps['preprocess'].get_feature_names_out()
    coefs = pipe.named_steps['reg'].regressor_.coef_
    coefs = np.ravel(coefs)
    coef_series = pd.Series(coefs, index=feature_names)
    top20 = coef_series.abs().sort_values(ascending=False).head(20).index
    print("\nTop 20 absolute coefficients (Ridge):")
    for name in top20:
        print(f"{name:40s} {coef_series[name]: .6f}")
except Exception as e:
    print("Skip coefficient summary:", e)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
import joblib
import pandas as pd

# โหลดโมเดลและคอลัมน์
model = joblib.load("house_price_pipeline.joblib")
train_cols = pd.read_csv("train_cleaned.csv").drop(columns=["SalePrice"]).columns

app = FastAPI(title="House Price Prediction API")

# สร้าง class สำหรับ input
class HouseInput(BaseModel):
    OverallQual: float
    GrLivArea: float
    GarageCars: float
    YearBuilt: float
    YearRemodAdd: float
    FullBath: float
    KitchenQual: str
    Neighborhood: str
    LotArea: float
    TotalBsmtSF: float
    FirstFlrSF: float
    SecondFlrSF: float
    Fireplaces: float
    MSZoning: str
    BldgType: str

@app.post("/predict")
def predict_price(data: HouseInput):
    # สร้าง DataFrame จาก input
    input_dict = data.dict()
    df = pd.DataFrame([{col: input_dict.get(col, None) for col in train_cols}])
    
    # ทำนายราคา
    pred = model.predict(df)[0]
    return {"predicted_price": round(float(pred), 2)}
```
